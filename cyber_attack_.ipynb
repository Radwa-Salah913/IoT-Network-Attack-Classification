**Split data to X&Y**

```python
x=df.drop(['attack_type'],axis=1)
y=df['attack_type']
```

```python
def handle_outliers(col_name):
  Q1=df[col_name].quantile(0.25)
  Q3=df[col_name].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  df[col_name][df[col_name]>Q3]=Q3
  df[col_name][df[col_name]<Q1]=Q1
```

```python
for name in x.columns:
  handle_outliers(name)
```

**Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(x) # shoud apply normalizatio on (X) only
x=pd.DataFrame(scaled_data,columns=x.columns)
```

**ckeck balance of classes**

```python
y.value_counts()   # use class_weight as hyper_parameter in model
```

# ML Models

**Split data to train & test**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**SVM**

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',decision_function_shape='ovr',class_weight='balanced')
svm.fit(x_train,y_train)
```

```python
y_pred=svm.predict(x_test)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_pred,y_test)
print(report)
```

**Random forest**

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=100, random_state=42,class_weight='balanced')
RF.fit(x_train,y_train)
```
